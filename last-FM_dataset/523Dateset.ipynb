```python
from zmq.constants import NULL
from typing_extensions import dataclass_transform
import json

# Opening JSON file
json_root = PATH_OF_DATA + '/lastfm_asia_features.json'
f = open(json_root)

# returns JSON object as
# a dictionary
data = json.load(f)

# find the largest number in all items
max_num = 0 
for i in range(len(data)):
  num = str(i)
  if data[num]:
    if max(data[num]) > max_num:
      max_num = max(data[num])

print(max_num)


# Closing file
# f.close()
```

```python
# recreate the new feature json file
new_dict = {}
empty_feature = []
for i in range(max_num+1):
  empty_feature.append(0)

for i in range(len(data)):
  num = str(i)
  # print(num)
  feature = empty_feature.copy()
  for item in data[num]:
    feature[item] = 1
  new_dict[num] = feature
print(new_dict["0"])

json_object = json.dumps(new_dict)
 
# Writing to sample.json
with open("/content/drive/MyDrive/523-dataset/processed_feature.json", "w") as outfile:
    outfile.write(json_object)
    
```

Some tests and visualizations of the datasets

```python
import csv
 
target_root = PATH_OF_DATA + '/lastfm_asia_target.csv'
 
location_dict = {}
with open(target_root, 'r') as data:
  for line in csv.DictReader(data):
      # print(line)
      id = line['id']
      location = line['target']
      location_dict[id] = location
   
print(location_dict)

location_info = {}
for i in range(18):
  id_list = []
  id = str(i)
  location_info[id] = id_list

print(location_info)

for id in location_dict:
  location_info[location_dict[id]].append(id)

print(location_info)



```

```python
import matplotlib.pyplot as plt

x = list(location_info.keys())
y = []
for id in location_info:
  y.append(len(location_info[id]))

print(y)

plt.bar(x,y)
plt.title("location count")
plt.xlabel("location id")
plt.ylabel('count')
plt.show()
```

```python
edge_root = PATH_OF_DATA + '/lastfm_asia_edges.csv'
 
edge_dict = {}
for i in range(7624):
  id = str(i)
  edge_list = []
  edge_dict[id] = edge_list

print(edge_dict)

with open(edge_root, 'r') as data:
  for line in csv.DictReader(data):
      # print(line)
      id = line['node_1']
      edge = line['node_2']
      edge_dict[id].append(edge)
   
print(edge_dict)
```

```python
differnt_location_count = {}
edge_count = []
dif_count_total = 0
total_edge = 0
for i in range(18):
  id = str(i)
  different_count = 0
  differnt_location_count[id] = different_count
  edge_count.append(0)

print(differnt_location_count)
print(edge_count)

for location in location_info:
  for user in location_info[location]:
    for edge in edge_dict[user]:
      num = int(location)
      edge_count[num] += 1
      total_edge += 1
      if location_dict[edge] != location:
        differnt_location_count[location] += 1
        dif_count_total += 1


print(differnt_location_count)
print(edge_count)
print(dif_count_total)
print(total_edge)
print(dif_count_total/total_edge)

```

```python
different_location_percentage = []
for id in differnt_location_count:
  num = int(id)
  percentage = differnt_location_count[id]/edge_count[num]
  # print(percentage)
  different_location_percentage.append(percentage)

plt.bar(x,different_location_percentage)
plt.title("different edge percentage")
plt.xlabel("location id")
plt.ylabel("edges to different location / all edges")
plt.show()
```

```python
plt.bar(x,edge_count, color ='r',width = 1,edgecolor ='grey', label ='all edges')
plt.bar(x,differnt_location_count.values(), color ='g',width = 1,edgecolor ='grey', label ='edges to different location')

plt.title("edges vs edges to differnt location")
plt.xlabel("location id")
plt.ylabel('number')
plt.legend()
plt.show()
```
